```python
# A2.3

#tasks 2 3 and 4( did visualization and normalization on the 49 participants that is on 49 rows)

import pandas as pd

# Load the data
df = pd.read_csv('aw_fb_data.csv')

# Creating a new column called "participant_id"
df['participant_id'] = df['age'].astype(str) + '_' + df['gender'].astype(str) + '_' + df['height'].astype(str) + '_' + df['weight'].astype(str)

# Keeping one sample from each participant
df_filtered = df.drop_duplicates(subset='participant_id')

#checking if the df contains 49 participants or not
num_participants = df_filtered.shape[0]
print(f'Total number of participants after dropping duplicates: {num_participants}')

# Visualizing the age, height, and weight
fig, axes = plt.subplots(3, 1, figsize=(10, 15))

# age
axes[0].plot(df_filtered['age'], color='red')
axes[0].set_title('Age')
axes[0].grid(True)
axes[0].legend(['Age'])

# height
axes[1].plot(df_filtered['height'], color='green')
axes[1].set_title('Height')
axes[1].grid(True)
axes[1].legend(['Height'])

# weight
axes[2].plot(df_filtered['weight'], color='blue')
axes[2].set_title('Weight')
axes[2].grid(True)
axes[2].legend(['Weight'])

plt.tight_layout()
plt.show()

# task 3

# first three participants
df_participant1 = df_filtered.head(1)
df_participant2 = df_filtered.iloc[1:3].reset_index(drop=True)
df_participant3 = df_filtered.iloc[3:].reset_index(drop=True)

# plotting
fig, axes = plt.subplots(3, 1, figsize=(10, 15), sharex=True)

# steps
axes[0].plot(df_participant1['steps'], color='red', label='Participant #1')
axes[0].plot(df_participant2['steps'], color='green', label='Participant #2')
axes[0].plot(df_participant3['steps'], color='blue', label='Participant #3')
axes[0].set_title('Steps')
axes[0].grid(True)
axes[0].legend(loc='upper left')

# heart rate
axes[1].plot(df_participant1['hear_rate'], color='red', label='Participant #1')
axes[1].plot(df_participant2['hear_rate'], color='green', label='Participant #2')
axes[1].plot(df_participant3['hear_rate'], color='blue', label='Participant #3')
axes[1].set_title('Heart Rate')
axes[1].grid(True)
axes[1].legend(loc='upper left')

# calories
axes[2].plot(df_participant1['calories'], color='red', label='Participant #1')
axes[2].plot(df_participant2['calories'], color='green', label='Participant #2')
axes[2].plot(df_participant3['calories'], color='blue', label='Participant #3')
axes[2].set_title('Calories')
axes[2].grid(True)
axes[2].legend(loc='upper left')

plt.tight_layout()
plt.show()


# task 4

# Normalizing (min-max) age, height, and weight
df_filtered["age_norm"] = (df_filtered["age"] - df_filtered["age"].min()) / (df_filtered["age"].max() - df_filtered["age"].min())
df_filtered["height_norm"] = (df_filtered["height"] - df_filtered["height"].min()) / (df_filtered["height"].max() - df_filtered["height"].min())
df_filtered["weight_norm"] = (df_filtered["weight"] - df_filtered["weight"].min()) / (df_filtered["weight"].max() - df_filtered["weight"].min())

# Standardizing (z-score) steps and heart_rate
df_filtered["steps_std"] = (df_filtered["steps"] - df_filtered["steps"].mean()) / df_filtered["steps"].std()
df_filtered["hear_rate_std"] = (df_filtered["hear_rate"] - df_filtered["hear_rate"].mean()) / df_filtered["hear_rate"].std()

print(df_filtered)
```
